## Paso 2. Haz informes y calcula métricas

## The happiest countries in the world are:

```python
happiest = happiness.head()
happiest
```

```python
happiest['GDP per capita'].plot(xlabel='Country or region',ylabel='GDP per capita', y='GDP per capita', kind='bar')
```

## The least happiest countries in the world

```python
least_happiest = happiness.tail()
least_happiest
```

```python
least_happiest['GDP per capita'].plot(xlabel='Country or region',ylabel='GDP per capita', y='GDP per capita', kind='bar')
```
